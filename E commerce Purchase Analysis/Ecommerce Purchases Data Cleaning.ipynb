### 14. How Many People have A Credit Card That Expires In 2020?

```python
def split_cc(df):
    day_expires = []
    year_expires = []
    for i in df:
        print(i)
        day_expires.append(int(i.split('/')[0]))
        year_expires.append(int(i.split('/')[1]))  
    return day_expires,year_expires
```

```python
day,year = split_cc(data["CC Exp Date"])
```

```python
data["CC Exp year"] = year
data["CC Exp day"] = day
data.head()
```

```python
data.loc[data["CC Exp year"] == 20].shape[0]
```

### Get Inforamtion in Browser Info

```python
data["Browser Info"].unique()[:10]
```

```python
def clean_browserInfo(browser_info):
    system_ = ["Linux",'Macintosh','Windows']
    value_system = browser_info.split('/')[1].split('(')[1]
    item = ''
    for i in system_:
        if i in value_system:
            item = i
            break
    return [browser_info.split('/')[0],item]
```

```python
browser_info = data["Browser Info"].apply(clean_browserInfo)
browser_info
```
